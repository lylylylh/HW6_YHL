```python
import csv
def func(data,k):
    next(data)
    next(data)
    row=next(data)
    i=1
    for v in row:
        if i%4==3:
            male[k]+=int(v)
        if i%4==0:
            female[k]+=int(v)
        i+=1
f1=open('jeju2008.csv','r')
f2=open('jeju2013.csv','r')
f3=open('jeju2018.csv','r')
f4=open('jeju2022.csv','r')
male=[0,0,0,0]
female=[0,0,0,0]
data=csv.reader(f1)
func(data,0)
data=csv.reader(f2)
func(data,1)
data=csv.reader(f3)
func(data,2)
data=csv.reader(f4)
func(data,3)
f1.close()
f2.close()
f3.close()
f4.close()
```

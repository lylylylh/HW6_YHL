```python
import csv
def func(data,k):
    next(data)
    next(data)
    row=next(data)
    i=1
    for v in row:
        if i%4==3:
            male[k]+=int(v)
        if i%4==0:
            female[k]+=int(v)
        i+=1
f1=open('jeju2008.csv','r')
f2=open('jeju2013.csv','r')
f3=open('jeju2018.csv','r')
f4=open('jeju2022.csv','r')
male=[0,0,0,0]
female=[0,0,0,0]
data=csv.reader(f1)
func(data,0)
data=csv.reader(f2)
func(data,1)
data=csv.reader(f3)
func(data,2)
data=csv.reader(f4)
func(data,3)
f1.close()
f2.close()
f3.close()
f4.close()
```

```python
import matplotlib.pyplot as plt
f,axes=plt.subplots(2,2)
f.set_size_inches(10,10)
plt.subplots_adjust(wspace=0.15,hspace=0.3)
label=['Male',"Female"]
color=['lightblue','lightpink']
axes[0,0].pie([male[0],female[0]],labels=label,autopct='%.1f%%',colors=color)
axes[0,0].set_title("2008", fontsize=15)
axes[0,0].legend(fontsize=9)
axes[0,1].pie([male[1],female[1]],labels=label,autopct='%.1f%%',colors=color)
axes[0,1].set_title("2013", fontsize=15)
axes[0,1].legend(fontsize=9)
axes[1,0].pie([male[2],female[2]],labels=label,autopct='%.1f%%',colors=color)
axes[1,0].set_title("2018", fontsize=15)
axes[1,0].legend(fontsize=9)
axes[1,1].pie([male[3],female[3]],labels=label,autopct='%.1f%%',colors=color)
axes[1,1].set_title("2022", fontsize=15)
axes[1,1].legend(fontsize=9)
plt.show()

print("\n* 제주도의 남녀 비율은")
print("2008년 남성:43.3% 여성:56.7%")
print("2013년 남성:44.4% 여성:55.6%")
print("2018년 남성:45.9% 여성:54.1%")
print("2022년 남성:47.7% 여성:52.3%")
print("으로 꾸준히 남성은 50% 이하이며 여성은 50% 이상을 차지하는 것을 볼 수 있다.")
print("그러므로 제주도에는 여성의 비율이 더 높다 라는 속설은 참이라는 것을 알 수 있다.")
```

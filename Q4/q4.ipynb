```python
import csv
f=open('subway2303.csv','r')
data=csv.reader(f)
next(data)
next(data)
on=dict()
off=dict()
oo=dict()
for row in data:
    b1=0
    b2=0
    b3=0
    onkeys=on.keys()
    offkeys=off.keys()
    ookeys=oo.keys()
    for i in range(4,8):
        row[i]=int(row[i])
    for k in onkeys:
        if(k==row[3]):
            on[k]+=row[4]+row[6]
            b1=1
    for k in offkeys:
        if(k==row[3]):
            off[k]+=row[5]+row[7]
            b2=1
    for k in ookeys:
        if(k==row[3]):
            oo[k]+=row[4]+row[5]+row[6]+row[7]
            b3=1
    if b1==0:
        on[row[3]]=row[4]+row[6]
    if b2==0:
        off[row[3]]=row[5]+row[7]
    if b3==0:
        oo[row[3]]=row[4]+row[5]+row[6]+row[7]
onv=list(on.values())
ofv=list(off.values())
oov=list(oo.values())
onv.sort(reverse=True)
ofv.sort(reverse=True)
oov.sort(reverse=True)
onv=onv[0:30]
ofv=ofv[0:30]
oov=oov[0:30]
onconvert={v:k for k,v in on.items()}
offconvert={v:k for k,v in off.items()}
ooconvert={v:k for k,v in oo.items()}
onk=[]
ofk=[]
ook=[]
for value in onv:
    onk.append(onconvert.get(value)) 
for value in ofv:
    ofk.append(offconvert.get(value)) 
for value in oov:
    ook.append(ooconvert.get(value)) 
f.close()
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='NanumGothic')

f,axes=plt.subplots(3,1,constrained_layout=True)
f.set_size_inches(8,25)
f.suptitle("             30 Station with Max Passengers on Rush Hour\n",fontsize=16)

axes[0].barh(onk,onv,color='pink')
axes[0].set_xlabel("Number of passengers") 
axes[0].set_ylabel("Station")
axes[0].set_title("< Get On >") 

axes[1].barh(ofk,ofv,color='gold')
axes[1].set_xlabel("Number of passengers") 
axes[1].set_ylabel("Station")
axes[1].set_title("< Get Off >") 

axes[2].barh(ook,oov,color='skyblue')
axes[2].set_xlabel("Number of passengers") 
axes[2].set_ylabel("Station")
axes[2].set_title("< Get On + Get Off >") 
plt.show()
print("\n* 최대 승차역의 승객수를 보면 50만 이상으로 압도적으로 높은 신림역을 제외하고 15만~35만 사이이다.")
print("최대 하차역의 승객수는 30위인 안국역부터 20만이 넘으며 50만이 넘는 역도 7개가 존재한다.")
print("이를 통해 최대 승차역들의 승객수보다 최대 하차역들의 승객수가 더 높은 것을 알 수 있다.")
print("출근 시간대의 표본이므로 출발지점은 도착지점보다 다양하고 하차하는 승객들은 오피스 타운인 도착지점으로")
print("몰리기 때문에 이러한 결과가 나온 것으로 추측해볼 수 있다.")
print("또한 하차역의 높은 순위권의 역들과 승하차역의 높은 순위권의 역들이 많이 겹치며")
print("두 자료의 1,2,3위의 집합은 같은 것을 알 수 있다.")
print("이 또한 앞서 언급한 추측에 해당하며 승하차역 순위는 승차역보다는 하차역의 양상과 비슷한 것을 볼 수 있다.")
```

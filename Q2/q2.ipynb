```python
import random
import csv
f=open('random.csv','w',newline='')
wr=csv.writer(f)
i=2
header=[]
header.append("시행횟수")
for k in range(0,4):
    r=10**i
    i+=1
    s=str(r)+"번"
    header.append(s)
wr.writerow(header)
arr1=[]
arr2=[]
arr3=[]
arr4=[]
for i in range(0,100):
    r=round(float(random.random()*5+1),2)
    arr1.append(r)
for i in range(0,1000):
    r=round(float(random.random()*5+1),2)
    arr2.append(r)
for i in range(0,10000):
    r=round(float(random.random()*5+1),2)
    arr3.append(r)
for i in range(0,100000):
    r=round(float(random.random()*5+1),2)
    arr4.append(r)
for i in range(0,100000):
    if i>=len(arr1):
        arr1.append('')
    if i>=len(arr2):
        arr2.append('')
    if i>=len(arr3):
        arr3.append('')
    wr.writerow([str(i+1)+"번째수",arr1[i],arr2[i],arr3[i],arr4[i]])
f.close()
```

```python
import matplotlib.pyplot as plt
f=open('random.csv','r')
data=csv.reader(f)
header=next(data)
d1=[]
d2=[]
d3=[]
d4=[]
for row in data:
    if row[1]!='':
        d1.append(float(row[1]))
    if row[2]!='':
        d2.append(float(row[2]))
    if row[3]!='':
        d3.append(float(row[3]))
    d4.append(float(row[4]))
f,axes=plt.subplots(2,2)
f.set_size_inches(15,10)
plt.subplots_adjust(wspace=0.15,hspace=0.3)

axes[0,0].hist(d1,bins=60,color='lightpink')
axes[0,0].set_xlabel('random number')
axes[0,0].set_ylabel('frequency')
axes[0,0].set_title("Random Number <1~6> x 100", fontsize=10)

axes[0,1].hist(d2,bins=60,color='mediumpurple')
axes[0,1].set_xlabel('random number')
axes[0,1].set_ylabel('frequency')
axes[0,1].set_title("Random Number <1~6> x 1000",fontsize=10)

axes[1,0].hist(d3,bins=60,color='lightskyblue')
axes[1,0].set_xlabel('random number')
axes[1,0].set_ylabel('frequency')
axes[1,0].set_title("Random Number <1~6> x 10000",fontsize=10)

axes[1,1].hist(d4,bins=60,color='gold')
axes[1,1].set_xlabel('random number')
axes[1,1].set_ylabel('frequency')
axes[1,1].set_title("Random Number <1~6> x 100000",fontsize=10)
plt.show()

print("\n위 4개의 히스토그램을 보면")
print("100,1000,10000,100000 순서대로 랜덤 숫자 생성을 시행하는 횟수가 증가할수록")
print("(같은 비율이라고 가정했을 때) 그래프의 y값들 간의 격차가 시각적으로 줄어드는 것을 볼 수 있다.")
print("그러므로 시행 결과가 점점 고르게 분포한다는 것을 알 수 있다.")
```
